```python
%%writefile wandb_mp.py
import multiprocessing as mp
import wandb
import os
import numpy as np

def do_work(emoji):
    emoji,name = emoji
    run = wandb.init(entity='demonstrations',project='ddp_training',group='ddp_different_runs_10')
    pid =  os.getpid()
    run.name = f'{name} with pid {pid}'
    print(f' {emoji} I am process {name} and my id is {pid}')
    mu , sigma = 0, 0.1
    loss, acc = np.linspace(1,0,50), np.linspace(0,1,50)
    loss = loss+np.random.normal(mu, sigma, 50)
    acc = acc+np.random.normal(mu, sigma, 50)
    for ls,ac in zip(loss,acc):
        to_log = { 'loss':ls, 'acc':ac}
        run.log(to_log)
    run.finish()

def get_time_emoji():
    '''generatory for team and runs'''
    amazing_team = [name.title() for name in 
                  ['mo','arti','luis','thanos',
                    'bill','leslie','nate','ramit',
                    'anmol','frida','raphael']]
    emoji = ['\U0001F561','\U0001F556','\U0001F562','\U0001F557',
             '\U0001F563','\U0001F558','\U0001F564','\U0001F561',
             '\U0001F559','\U0001F565','\U0001F566','\U0001F555']
    
    for emogi in zip(emoji,amazing_team):
        yield emogi
          
def main():
    print(f'i am main and my id is {os.getpid()}')
    gen = get_time_emoji()
    emo = get_time_emoji
    wandb.setup()
    pool = mp.Pool(processes=10)
    pool.map(do_work,gen)

if __name__ == "__main__":
    main()
```

```python
%%writefile wandb_mp_2.py
import multiprocessing as mp
import wandb
import os
import numpy as np

def train(run,emoji):
    emoji,name = emoji
    run = wandb.init(entity='demonstrations',project='ddp_training',group='ddp_different_runs')
    run.name 
    pid =  os.getpid()
    run.name = str(pid)
    print(f' {emoji} I am process {name} and my id is {pid}')
    mu , sigma = 0, 0.1
    loss, acc = np.linspace(1,0,50), np.linspace(0,1,50)
    loss = loss+np.random.normal(mu, sigma, 50)
    acc = acc+np.random.normal(mu, sigma, 50)
    for ls,ac in zip(loss,acc):
        to_log = { 'loss':ls, 'acc':ac}
        run.log(to_log)
    run.finish()
    
def get_time_emoji():
    '''generatory for team and runs'''
    amazing_team = ['mo','arti','luis','thanos',
                    'bill','leslie','nate','ramit',
                    'anmol','frida','raphael']
    emoji = ['\U0001F561','\U0001F556','\U0001F562','\U0001F557',
             '\U0001F563','\U0001F558','\U0001F564','\U0001F561',
             '\U0001F559','\U0001F565','\U0001F566','\U0001F555']
    return (emoji, amazing_team)

def main():
    print(f'i am main and my id is {os.getpid()}')
    run = wandb.init(entity='demonstrations',project='ddp_training',group='same_run')
    emo = get_time_emoji()
    p = mp.Process(target=train, kwargs=dict(run=run,emoji=emo))
    p.start()
    p.join()
        
if __name__ == "__main__":
    main()
```

```python
%%writefile wandb_sweep_mp.py
import multiprocessing as mp
import wandb
import os
import numpy as np

sweep_configuration = {
    "name": "my-awesome-sweep",
    "metric": {"name": "accuracy", "goal": "maximize"},
    "method": "grid",
    "parameters": {
        "a": {
            "values": [1, 2, 3, 4,5,6,7,8,9]
        },
        "b": {
            "values": [1, 2, 3, 4,5,6,7,8,9]
        },
        "c": {
            "values": [1, 2, 3, 4,5,6,7,8,9]
        }
    }
}

def sub_sweep(swp_id):
    def do_work(run=None):
        run = wandb.init()
        pid =  os.getpid()
        mu , sigma = 0, 0.1
        loss, acc = np.linspace(1,0,50), np.linspace(0,1,50)
        loss = loss+np.random.normal(mu, sigma, 50)
        acc = acc+np.random.normal(mu, sigma, 50)
        for ls,ac in zip(loss,acc):
            to_log = { 'loss':ls, 'acc':ac}
            run.log(to_log)
        run.finish()
    wandb.agent(swp_id, function=do_work,count=10)

def main():
    n_processes = 20
    sweep_id = wandb.sweep(sweep_configuration,entity='demonstrations',project='spawn_sweep')
    sweep_id = [sweep_id for p in range(n_processes)]
    print(f'i am main and my id is {os.getpid()}')
    wandb.setup()
    pool = mp.Pool(processes= n_processes)
    pool.map(sub_sweep,sweep_id)

if __name__ == "__main__":
    main()
```
